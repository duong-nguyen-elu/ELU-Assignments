```python
#Exporting current_weather data with API call from OpenWeatherMap site
import requests
import json
import pandas as pd

#Defining the jprint func to print out json output as formatted string
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)

url = "http://api.openweathermap.org/data/2.5/weather?id=6544259&appid=6c80b405e6aea2f6e872dfc3a94fe4b9&lang=en&units=metric"
current_weather = requests.get(url).json()
jprint(current_weather)
```

```python
#As the output is having mixed dicts type with other strings, I modify it to all strings output for csv output
current_weather['clouds_all'] = 100
current_weather['coord_lat'] = 52.3686
current_weather['coord_lon'] = 5.2375
current_weather['feels_like'] = -1.25
current_weather['humidity'] = 88
current_weather['pressure'] = 998
current_weather['temp'] = 1.14
current_weather['temp_max'] = 2.22
current_weather['temp_min'] = 0.56
current_weather['country'] = "NL"
current_weather['city_id'] = 2036661
current_weather['sunrise'] = 1611473504
current_weather['sunset'] = 1611504617
current_weather['type'] = 3
current_weather['weather_description'] = "overcast clouds"
current_weather['weather_icon'] = "04n"
current_weather['weather_id'] = 804
current_weather['weather_main'] = "Clouds"
current_weather['wind_deg'] = 95
current_weather['wind_gust'] = 1.04
current_weather['wind_speed'] = 0.45

del(current_weather['clouds'])
del(current_weather['coord'])
del(current_weather['main'])
del(current_weather['sys'])
del(current_weather['weather'])
del(current_weather['wind'])

jprint(current_weather)
```

```python
with open('current_weather.json', 'w') as outfile:
    json.dump(current_weather, outfile)
```

```python
#Exporting 5-day / 3-hour forecast data with API call from OpenWeatherMap site
import requests
url = "http://api.openweathermap.org/data/2.5/forecast?id=6544259&appid=6c80b405e6aea2f6e872dfc3a94fe4b9&lang=en&units=metric"
weather_forecast = requests.get(url).json()
df = pd.json_normalize(weather_forecast['list'])
df
```
