```python
import pandas as pd
import numpy as np
```

```python
app_train = pd.read_csv('credit-risk/application_train.csv')
previous_app = pd.read_csv('credit-risk/previous_application.csv')
credit_balance = pd.read_csv('credit-risk/credit_card_balance.csv')
```

```python
#calculate the average loan size for all loan applicants from the application_train table
avg_loan_size = app_train[['AMT_CREDIT']].mean()[0]
print(f'The average loan size for all loan applicants from the application_train table is {avg_loan_size}')
```

```python
#calculate how many applicants from the application_train table were previously rejected
full_app = app_train[['SK_ID_CURR','CODE_GENDER','CNT_CHILDREN','DAYS_BIRTH']].merge(previous_app[['SK_ID_CURR', 'NAME_CONTRACT_STATUS']], how='inner', on='SK_ID_CURR')
num_rejected = full_app[full_app['NAME_CONTRACT_STATUS'] == 'Refused'].SK_ID_CURR.nunique()
print(f'The number of applicants from the application_train table were previously rejected is {num_rejected}')
```

```python
#calculate the average credit card balance for the applicants from the application_train table
credit_app = app_train[['SK_ID_CURR']].merge(credit_balance[['SK_ID_CURR', 'AMT_BALANCE', 'NAME_CONTRACT_STATUS']], how='inner', on='SK_ID_CURR')
app_credit_balance = credit_app.AMT_BALANCE.mean()
print(f'The average credit card balance for the applicants from the application_train table is {app_credit_balance}')
credit_app.groupby('NAME_CONTRACT_STATUS').AMT_BALANCE.mean()
```

```python
#What is the ratio of approved/refused loans per gender?
gender_encoding = pd.get_dummies(full_app.CODE_GENDER, prefix='Gender')
full_app_encoding = pd.concat([full_app, gender_encoding], axis=1)
full_app_encoding.groupby('NAME_CONTRACT_STATUS')['Gender_F','Gender_M','Gender_XNA'].mean()
```

```python
#Do people with children have a higher or lower loan approval ratio than those without?
no_child_approved_perc = full_app[(full_app['CNT_CHILDREN'] == 0) & (full_app['NAME_CONTRACT_STATUS'] == 'Approved')].SK_ID_CURR.nunique()/full_app[full_app['CNT_CHILDREN'] == 0].SK_ID_CURR.nunique()
has_child_approved_perc = full_app[(full_app['CNT_CHILDREN'] > 0) & (full_app['NAME_CONTRACT_STATUS'] == 'Approved')].SK_ID_CURR.nunique()/full_app[full_app['CNT_CHILDREN'] > 0].SK_ID_CURR.nunique()

if no_child_approved_perc > has_child_approved_perc:
    print('People has no child has higher loan approval ratio than people with children.')
elif no_child_approved_perc == has_child_approved_perc:
    print('Regardless of number of children, people has same loan approval ratio.')
else:
    print('People has no child has less loan approval ratio than people with children.')
```

```python
#Bin client ages in young, middle age and senior citizens -- and create a pivot table showing percentages of loans Approved, Cancelled and Refused per age group.
df_new = full_app.assign(Age = lambda full_app: (full_app.loc[:, 'DAYS_BIRTH'].abs())//365.2425)
#full_app['age'] = [((age_in_days).abs()//365.2425) for age_in_days in full_app[['DAYS_BIRTH']]]
#full_app.head()
df_new.head()

bin_labels_3 = ['Young', 'Middle_Age', 'Senior_Citizens']
df_new['Age_Group'] = pd.qcut(df_new['Age'],
                              q=[0, .33, .67, 1],
                              labels=bin_labels_3)
df_new.head()

age_encoding = pd.get_dummies(df_new.Age_Group, prefix='Group')
full_app_encoding = pd.concat([df_new, age_encoding], axis=1)
full_app_encoding.groupby('NAME_CONTRACT_STATUS')['Group_Young','Group_Middle_Age','Group_Senior_Citizens'].mean()
```

```python
#Find clients that have the most recorded loans in the Credit bureau (above 95th percentile). What percentage of those have had at least one loan with STATUS == 5 (DPD 120+) in their Credit Bureau history?
bureau_credit = pd.read_csv('credit-risk/bureau.csv')
bureau_credit.head()

top_bureau_credit_amt = bureau_credit[bureau_credit.AMT_CREDIT_SUM > bureau_credit.AMT_CREDIT_SUM.quantile(.95)]
top_bureau_credit_amt
```

```python
bureau_balance = pd.read_csv('credit-risk/bureau_balance.csv')
bureau_balance.head()

bureau_join = bureau_credit[['SK_ID_CURR','SK_ID_BUREAU']].merge(bureau_balance[['SK_ID_BUREAU', 'STATUS']], how='inner', on='SK_ID_BUREAU')

overdue = (bureau_join[bureau_join['STATUS'] == '5'].SK_ID_CURR.nunique()/bureau_join.SK_ID_CURR.nunique())*100

print(f'The percentage of those have had at least one loan with STATUS == 5 (DPD 120+) in their Credit Bureau history is {overdue}%')
```
